### **1. Definición del problema a resolver y enfoque elegido para ello.**

El problema gira entorno a la obtención de energía renovable a partir de las olas. El medio marino es una fuente de energía poco explotada para la que se han ideado múltiples dispositivos de extracción de energía renovable. Nuestro problema, en particular, emplea unos dispositivos WECs (Wave Energy Converters) denominados CETO.
Los CETO son unos dispositivos situados en el fondo marino que están sujetos a una bomba con la que son capaces de transmitir, a través de unas tuberías, agua a gran presión hasta una central, donde se convertirá en energía eléctrica gracias a una turbina.

#### **Dataset**
La base de datos de este proyecto la podemos encontrar en https://archive.ics.uci.edu/ml/datasets/Wave+Energy+Converters. Analizaremos el problema correspondiente a esta base de datos y aplicaremos técnicas de Aprendizaje Automático para resolverlo. La base de datos contiene información sobre la posición y potencia absorbida por los WECs en cuatro lugares distintos de la costa australiana: Sydney, Adelaide, Perth y Tasmania. Los WECs serán dispositivos CETO y habŕa colocados 16 de ellos en un área restringida de 566m2. El conjunto de datos pretende servir para estimar un modelo que sea capaz de predecir la potencia total (Watts) absorbida por los 16 CETOs. La base de datos cuenta con 288,000 instancias, donde 72,000 aproximadamente corresponden a cada una de las zonas examinadas.

Los atributos que se incluyen en cada una de las bases de datos aportadas son: las posiciones de los CETOs en un área de 566m2 en el fondo marino de la zona indicada por la base de datos, las potencias absorbidas por cada CETO y la potencia total absorbida entre todos ellos, que se corresponde con la suma de las potencias individuales. Siendo $(x_i,y_i)$ las coordenadas de posicionamiento del CETO $i$−ésimo y $P_i$ su potencia, las características están ordenadas de la siguiente forma: $(x_1,..., x_{16},y_1,...,y_{16},P_1,...,P_{16},P_T)$, donde $P_T = \sum_{i=1}^{16}P_i$ es la potencia total absorbida.

Para nuestra tarea de predecir la potencia total (Watts), eliminaremos las potencias intermedias $P_i$, $i \in \{1,...,16\}$, para estimar directamente $P_T$ . A la hora de predecir la potencia real de una distribución de CETOs, sólo tendríamos sus posiciones. Así, dispondremos de 32 características (16 pares de coordenadas) y de un valor a estimar: potencia total (Watts).

#### **Comprensión del problema**

Nuestros datos saldrán de entre los datos de Sydney, Adelaide, Perth y Tasmania. Para resolver nuestro problema no podremos coger todos los conjuntos de datos y unirlos en uno solo debido a que, en realidad, los datos no están idénticamente distribuidos. Indudablemente, las regiones cuadradas de 566m2 no se encuentran en la misma localización submarina. Puede haber muchos factores que puedan afectar a la potencia total generada en cada una de estas localizaciones: profundidad y accidentes geográficos del fondo marino, corrientes marinas en la zona, fauna marina, etcétera. Bajo este fuerte indicio, decidimos comprobar las medias de las potencias generadas por cada configuración de CETOs recogidas en cada instancia de cada conjunto de datos (media de las potencias totales, $P_T$, en los conjuntos de entrenamiento de cada zona), obteniendo:

- Sydney: 1,486,224.
- Adelaide: 1,410,060.
- Perth: 1,394,507.
- Tasmania: 3,760,088.

Vemos que las diferencias entre las medias de las potencias totales (valor a predecir) existente entre los conjuntos de datos en función de la zona corroboran nuestra hipótesis. Por tanto, asumiremos que los datos de cada conjunto no siguen una misma distribución y por lo tanto no podemos considerar un único problema de aprendizaje para el conjunto de los 288,000 datos. Debido a ello, utilizaremos para nuestro problema únicamente el conjunto de datos de Sydney, y por lo tanto los resultados obtenidos sólo serán aplicables a esta zona.

Tras haber explicado conceptualmente el problema a resolver, pasamos ahora a traducirlo a notación de un problema de Aprendizaje Automático típico. Nuestro espacio de características será $\mathcal{X} =[0,566]^{32}$, que representa los 16 pares $(x,y)$ para la posición de cada CETO. Nuestra característica a predecir es $\mathcal{Y} = \mathbb{R} $, correspondiente a la potencia total en vatios. El objetivo es por tanto conseguir un modelo que sea capaz de encontrar una hipótesis $\mathcal{g}$ que se aproxime lo máximo posible a $f : \mathcal{X} → \mathcal{Y}$, que es la función verdadera que asigna un valor de $\mathcal{y}\in\mathcal{Y}$ a cada $\mathcal{x}\in\mathcal{X}$.

```python
# Clase para la generación de característica de distancia
# mínima entre cada dos CETOs en cada instancia
class DistanciaMinima:
    def __init__(self):
        pass
    # Calculamos las características
    def fit(self,X,y):
        self.mindist = np.zeros(X.shape[0])         # Distancia mínima entre cada dos CETOs para cada instancia
        for k in range(X.shape[0]):                 # Ahora en cada instancia
          dist = []
          for i in range(16):                       # Calculamos las distancias entre dos CETOs
            for j in range(i+1,16):
              dist.append(norm([(X[k,i] - X[k,j]), (X[k,i+16] - X[k,j+16])]))
          self.mindist[k] = np.min(np.array(dist))  # Añadimos la distancia mínima
        return self
    # Añadimos la característica a los datos
    def transform(self,X):
        return np.c_[X, self.mindist]

# Clase para la generación de característica de distancia
# máxima entre cada dos CETOs en cada instancia
class DistanciaMaxima:
    def __init__(self):
        pass
    # Cálculo de la característica a añadir
    def fit(self,X,y):
        self.maxdist = np.zeros(X.shape[0])        # Distancia máxima entre cada dos CETOs para cada instancia
        for k in range(X.shape[0]):                # Ahora en cada instancia
          dist = np.zeros((X.shape[1], X.shape[1]))
          for i in range(16):
            for j in range(i+1,16):
              dist[i,j] = norm([(X[k,i] - X[k,j]), (X[k,i+16] - X[k,j+16])])  # Distancias entre dos CETOs
          self.maxdist[k] = np.max(dist)           # Añadimos la distancia máxima
        return self
    # Añadimos la característica a los datos
    def transform(self,X):
        return np.c_[X, self.maxdist]

# Clase para la generación de característica de la media de las
# distancias mínimas entre cada dos CETOs en cada instancia
class MediaDistanciasMinimas:
    def __init__(self):
        pass
    # Cálculo de la característica a añadir
    def fit(self,X,y):
        self.mindist = np.zeros(X.shape[0])       # Donde guardamos las características a añadir
        for k in range(X.shape[0]):               # En cada instancia
          mindists = np.zeros(16)
          for i in range(16):                     # En cada CETO en dicha instancia
            min = float('inf')
            for j in range(16):
              curdist = norm([(X[k,i] - X[k,j]), (X[k,i+16] - X[k,j+16])])
              if curdist < min and not i == j:
                min = curdist                     # Calculamos la distancia mínima del CETO actual a cualquier otro
            mindists[i] = min
          self.mindist[k] = np.mean(mindists)     # Añadimos la media de las distancias mínimas calculadas
        return self
    # Añadimos la característica a los datos
    def transform(self,X):
        return np.c_[X, self.mindist]

# Clase para la generación de característica de la media de las
# distancias máximas entre cada dos CETOs en cada instancia
class MediaDistanciasMaximas:
    def __init__(self):
        pass
    # Cálculo de la característica a añadir
    def fit(self,X,y):
        self.maxdist = np.zeros(X.shape[0])       # Donde guardamos las características a añadir
        for k in range(X.shape[0]):               # En cada instancia
          maxdists = np.zeros(16)
          for i in range(16):                     # En cada CETO en dicha instancia
            max = 0
            for j in range(16):
              curdist = norm([(X[k,i] - X[k,j]), (X[k,i+16] - X[k,j+16])])
              if curdist > max and not i == j:
                max = curdist                     # Calculamos la distancia máxima del CETO actual a cualquier otro
            maxdists[i] = max
          self.maxdist[k] = np.mean(maxdists)     # Añadimos la media de las distancias máximas calculadas
        return self
    # Añadimos característica a los datos
    def transform(self,X):
        return np.c_[X, self.maxdist]

# Clase para el cálculo de la característica de la
# distancia media al centroide de los puntos
class DistanciaMediaACentroide:
    def __init__(self):
        pass
    #Cálculo de la característica a añadir
    def fit(self,X,y):
        self.distcent = np.zeros(X.shape[0])      # Donde guardamos las características a añadir
        for k in range(X.shape[0]):               # En cada instancia
          # Calculamos el centroide
          x_centr = np.mean(X[k,0:16])
          y_centr = np.mean(X[k,16:32])
          dists = np.zeros(16)
          for i in range(16):
            dists[i] = norm([(X[k,i]-x_centr), (X[k,i+16]-y_centr)])  # Distancia de cada CETO al centroide
          self.distcent[k] = np.mean(dists)       # Media de las distancias
        return self
    #Método para aplicar la transformación calculada.
    def transform(self,X):
        return np.c_[X, self.distcent]

# Clase para el cálculo de la característica de la varianza en las coordenadas de los CETOs
class Varianza:
    def init(self):
        pass
    #Cálculo de la característica a añadir
    def fit(self,X,y):
        self.vars = np.zeros(X.shape[0])       # Donde guardamos las características a añadir
        for k in range(X.shape[0]):              # Ahora en cada instancia
            # Codificamos los datos en forma de pares [x,y]
            pares = []
            for i in range(16):
                pares.append([X[k,i], X[k,i+16]])
            self.vars[k] = np.var(pares)      # Varianza de las coordeanadas
        return self
    #Método para aplicar la transformación calculada.
    def transform(self,X):
        return np.c_[X, self.vars]

# Clase encargada de generar una característica mediante la realización de un test
# estadístico de uniformidad Chi^2 sobre las coordenadas X de cada CETO en una instancia.
# La hipótesis nula [H0] sostiene que las coordenadas X de los CETOs
# exhiben una distribución uniforme en relación a las posiciones X.

# Clase que representa el p-value del test Chi cuadrado
class Chi2x:
    def init(self):
        pass
    # Método con el que encontramos las variables
    # con las que nos vamos a quedar
    def fit(self,X,y):
        self.p = np.zeros(X.shape[0])
        for k in range(X.shape[0]):
          freq = np.zeros(3)
          for i in range(16):
            if X[k][i]<566/3:
              freq[0] = freq[0]+1
            elif X[k][i]<2*566/3:
              freq[1] = freq[1]+1
            else:
              freq[2] = freq[2]+1
          chisq, p2 = chisquare(freq) # f_exp. Por defecto asumimos que las frecuencias equivalen
          self.p[k] = p2
        return self
    def transform(self,X):
        return np.c_[X, self.p]

# Clase que representa el p-value del test Chi cuadrado
# con una distribución Uniforme sobre las posiciones Y
class Chi2y:
    def init(self):
        pass
    # Método con el que encontramos las variables
    # con las que nos vamos a quedar
    def fit(self,X,y):
        self.p = np.zeros(X.shape[0])
        for k in range(X.shape[0]):
          freq = np.zeros(3)
          for i in range(16,32):
            if X[k][i]<566/3:
              freq[0] = freq[0]+1
            elif X[k][i]<2*566/3:
              freq[1] = freq[1]+1
            else:
              freq[2] = freq[2]+1
          chisq, p2 = chisquare(freq) # f_exp. Por defecto asumimos que las frecuencias equivalen
          self.p[k] = p2
        return self
    def transform(self,X):
        return np.c_[X, self.p]

# Clase que representa el p-value del test Chi
# cuadrado con una distribución Uniforme
class Chi2Diagonal:
    def init(self):
        pass
    # Método con el que encontramos las variables
    # con las que nos vamos a quedar
    def fit(self,X,y):
        self.p = np.zeros(X.shape[0])
        for k in range(X.shape[0]):     # En cada instancia
          freq = np.zeros(3)
          for i in range(16):           # En cada CETO en dicha instancia
            # Le asignamos una clase en función de su posición
            if (X[k][i]<2*566/3 and X[k][i+16]>566/3) and not (X[k][i]>566/3 and X[k][i+16]<2*566/3):
              freq[0] = freq[0]+1
            elif (X[k][i]>566/3 and X[k][i+16]<2*566/3) and not (X[k][i]<2*566/3 and X[k][i+16]>566/3):
              freq[1] = freq[1]+1
            else:
              freq[2] = freq[2]+1
          chisq, p2 = chisquare(freq)   # f_exp. Por defecto asumimos que las frecuencias equivalen
          self.p[k] = p2
        return self
    def transform(self,X):
        return np.c_[X, self.p]
```

##### **Ridge - Regresión Lineal.**

Minimiza la función objetivo $||y - Xw||^2_2 + α||w||^2_2$. Este modelo resuelve un modelo de regresión donde la función de pérdida es la función lineal de mínimos cuadrados y la regularización viene dada por la norma $\texttt{L2}$.

##### **Perceptrón Multicapa.**

Este modelo optimiza el error cuadrático utilizando SGD (el solver SGD lo hemos elegido nosotros entre las distintas opciones. Analizaremos más adelante el porqué).

##### **Support Vector Regressor.**

Utiliza una función de pérdida conocida como **Epsilon-Support Vector Regression Loss** o **Epsilon-Insensitive Loss**. Esta función de pérdida minimiza la diferencia entre las predicciones del modelo y los valores reales de los datos de entrenamiento, hasta un cierto límite ($ϵ$). Se define de la siguiente manera: $L(y, \hat{y}) = max(0, |y - \hat{y}| - ϵ)$, donde $y$ es el valor verdadero, $\hat{y}$ es la predicción del modelo y $\epsilon$ es el límite o umbral, también conocido como *epsilon-insensitive zone* (explicaremos este parámetro en la siguiente sección).

La función de pérdida se encarga de penalizar las diferencias entre las predicciones del modelo y los valores reales, a menos que la diferencia sea menor o igual a épsilon. Si la diferencia supera épsilon, la función de pérdida asigna un valor proporcional a la magnitud de la diferencia. Si la diferencia es menor o igual a épsilon, la función de pérdida asigna un valor de cero, ya que se considera que está dentro de la zona insensible a errores.

##### **AdaBoost**

Nosotros fijaremos la función de pérdida **linear loss** en el contexto de regresión. Esta función de pérdida se utiliza para medir la diferencia en valor absoluto entre las predicciones del modelo y los valores reales de los datos de entrenamiento: $L(y,\hat{y})=|y-\hat{y}|$.

##### **Random Forest**

Se ha fijado el parámetro $\texttt{criterion}$ (lo veremos) como el error cuadrático medio, lo que implica la reducción de la varianza como criterio de selección de características, minimizando la función de pérdida $\texttt{L2}$ a partir de la media de cada nodo terminal (info. extraída de https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html).

#### **Métricas de error**

Para el problema de regresión utilizaremos dos métricas distintas, siendo una de ellas el **Error Cuadrático Medio, ECM** (*Mean Squared Error*, MSE), y la otra el **coeficiente de determinación**, $R^2$.

El MSE de un estimador mide el promedio de los errores al cuadrado, es decir, la diferencia entre lo estimado y lo que se estima:

\begin{equation}
MSE(h) = \frac{1}{N}\sum_{i=0}^{N}(h(x_n)-y_n)^2,
\end{equation}
donde $h$ es el estimador.

El MSE es un error típico utilizado para la regresión lineal en el campo de la estadística, y supone una métrica de error estándar para los problemas de regresión en el campo del Aprendizaje Automático. Esta medida penaliza en gran medida los valores en los que la predicción ha sido muy errada. Está acotada inferiormente por 0, pero no superiormente. El MSE es el segundo momento (sobre el origen) del error, y por lo tanto incorpora tanto la varianza del estimador como su sesgo. Para un estimador insesgado, el MSE es la varianza del estimador.

Otras métricas auxiliares que se han empleado para medir la calidad de la solución obtenida son:

1. Al igual que la varianza, el MSE tiene las mismas unidades de medida que el cuadrado de la cantidad que se estima. Análogamente a lo que sucede con la desviación estándar, tomando la raíz cuadrada del MSE se obtiene lo que se denomina la desviación de la raíz cuadrada media, RMSE, que presenta las mismas unidades que la cantidad que se estima. Por lo tanto, para un estimador insesgado, el RMSE es la desviación estándar.

\begin{equation}
RMSE(h) = \sqrt{MSE(h)} = \sqrt{\frac{1}{N}\sum_{i=0}^{N}(h(x_n)-y_n)^2}.
\end{equation}

2. Además de esta métrica, utilizaremos otra complementaria que nos ayudará a interpretar de otra forma los errores cometidos. Es una métrica acotada superiormente en la que cuanto más alto es el valor resultante mejor es el modelo considerado. Nos referimos al coeficiente de determinación, denotado como $R^2$: \begin{equation}
R^2(h)=1-\frac{MSE(h)}{Var(y)}.
\end{equation} El denominador de la anterior fracción se corresponde con la varianza total de los datos, y el numerador con la varianza de los residuos, que coincide con el MSE (la división por el número de datos se puede simplificar). En el mejor de los casos, si la predicción es perfecta, la varianza de los residuos será 0, por lo que la medida valdrá 1. De igual forma, esta medida no está acotada inferiormente porque la varianza de los residuos (MSE) podría crecer ilimitadamente. Un valor $R^2(h) = 0$ significaría que nuestra función $h$ ha cometido un error igual a si se realizara una predicción constante igual a la media de los valores a predecir. Si el valor obtenido es menor que 0, la predicción sería en general peor que este caso; si es mayor que 0, obtenemos menos error en nuestra regresión de forma relativa a la varianza del modelo.

### **6. Argumentos a favor de la idoneidad de los modelos seleccionados para la BBDD.**

De cara a encontrar una buena aproximación de la función real $f$, vamos a utilizar diferentes clases de funciones que aporten distintos enfoques. En principio desconocemos la naturaleza de dicha función, por lo que podría tener sentido probar con diferentes clases que implementen diferentes tipos de aproximaciones.

Vamos a utilizar 5 clases de funciones diferentes, las cuales están asociadas a los siguientes modelos: Regresión Lineal, Perceptrón Multicapa, Máquinas de Vectores de Soporte (SVM), Random Forest y Redes de Funciones de base Radial. Explicaremos cada uno con más detalle a continuación.

#### **Clases lineales**

La clase de funciones lineales supone una clase de funciones muy simple y con una interpretabilidad natural. Por ello, si un modelo lineal proporciona una buena aproximación de la función real f, será preferible frente a cualquier otra que posea una bondad equiparable.
Si poseemos una serie de atributos $x = (1,x_1,...,x_n)$, la clase de funciones lineales sobre dichos atributos se define como: \begin{equation}
\mathcal{H}_{lin} = \{αx^T:α \in \mathbb{R}^{n+1}\}.\end{equation}

Podemos identificar la clase de funciones $\mathcal{H}_{lin} ≅ \mathbb{R}^{n+1}$, con lo que podremos hablar indistintamente tanto de funciones lineales como de vectores en $\mathbb{R}^{n+1}$ o, equivalentemente hiperplanos en el mismo espacio.

Esta clase de funciones tiene, sin embargo, claras limitaciones. Resulta bastante presuntuoso suponer que el fenómeno se corresponde con una simple separación lineal de las características consideradas sobre los datos, de forma que es bastante probable que el error por sesgo de estos modelos sea bastante grande. Surge entonces la necesidad de utilizar otros modelos más sofisticados capaces de predecir mejor la verdadera función $f$.

#### **Clases no lineales**

##### **Perceptrón Multicapa**
El Perceptrón Multicapa supone un intento de mejora directo de los modelos lineales. Se fundamenta en una aplicación consecutiva de funciones lineales y no lineales sobre las salidas de las anteriores a través de una estructura de capas, incrementando así el tamaño de la clase de funciones. La definición general de la clase de funciones definida por el Perceptrón Multicapa sobre un conjunto de atributos $x = (1, x_1, ..., x_n)$ es: \begin{equation} \mathcal{H}_{MLP} = \{W_L·θ(W_{L−1}·θ(...θ(W_1·x^T)...), W_i \in \mathcal{M}_{n_i,m_i} : m_i =n_{i−1}+1,i\in\{2,..,L\}\}, \end{equation}
donde $L\in \mathbb{N}$ y la función θ será una función no lineal. Tras realizar el producto por cada vector de pesos, se añade a las variables un nuevo término constantemente igual a 1. En nuestro caso, será $L = 3$, y además ocurrirá también que $n_3 = 1$.

El motivo por el que se ha elegido esta clase de funciones es por su gran versatilidad y gran capacidad de adaptación. Es una técnica que promete llegar a muy buenos resultados con una buena definición de la estructura del Perceptrón Multicapa. Sin embargo, el principal problema de estas funciones es que su potencial de adaptación depende de una forma muy fuerte de la arquitectura empleada (es decir, de los valores $n_i : i \in \{1,...,L\}$ y $L$ escogidos), y no tenemos un criterio claramente definido con el cual fijarlas para cada problema. Además, son modelos que tienden al sobreajuste muy rápidamente en base a estos parámetros y otros relativos al aprendizaje. Por tanto, será interesante escoger también clases de funciones que no dependan de una arquitectura prefijada por nosotros.

##### **Máquinas de Vectores de Soporte (SVM)**

En problemas de regresión como el nuestro puede emplearse la función **Support Vector Regressor** de *Scikit-Learn*. De la misma forma que para el problema de clasificación teníamos una suerte de *tubería* que definía la máxima separación entre clases, ahora contamos con un pasillo que, siendo lo más fino posible, debe contener el máximo número posible de puntos.

Se ha decidido emplear este modelo por las numerosas ventajas que proporciona. Es efectivo en espacios de alta dimensionalidad, es muy eficiente al emplear únicamente un subconjunto de los puntos de entrenamiento para la definición de la función, llamados vectores de soporte, y cuenta con una gran versatilidad gracias al uso de kernels. Todo esto hace lo convierte en un modelo muy interesante a tener en cuenta para nuestro problema.

Emplearemos un kernel de generación de características cúbicas, lo que nos permite aumentar inicialmente la dimensionalidad de los datos de una forma considerable, pero siempre siendo consecuentes con los recursos computacionales que tenemos disponibles, estableciendo un compromiso. Por consiguiente, la clase inicial de funciones que podremos aprender será realmente una clase de funciones cúbicas respecto al vector de características, con lo que, notando
$x = (1, x_1, ..., x_n,x_1^2, ...x_n^2, x_1x_2, ..., x_1x_n, x_2x_3, ...x_{n−1}x_n, x_1^3, ...x_n^3, x_1^2x_2, ..., x_{n−1}x_n^2)$,  nuestra clase de funciones será:
\begin{equation}
\mathcal{H}_{cub} = \{α^Tx : α \in \mathbb{R}^{1+3n+ \frac{3n(n−1)}{2} + \frac{n(n−1)(n−2)}{3}}\}. \end{equation}

Esta clase de funciones nos puede dar el regresor óptimo respecto a las características cúbicas respecto de las iniciales empleadas. Sin embargo, puede que esta aproximación sea excesivamente compleja y poco interpretable, y existan otras aproximaciones o paradigmas de clases de funciones y modelos de aprendizaje mucho más simples, interpretables y con un sesgo menor para la verdadera función $f$, la cual no tiene por qué tener una expresión analítica sencilla respecto a ningún conjunto de características polinómicas de las de partida.

##### **Boosting**
Un enfoque distinto a lo que hemos estado considerando hasta ahora es el del *ensemble learning*. El *ensemble learning* sostiene entrenar varios modelos y utilizar como valor predicho una combinación de los valores individuales devueltos por cada uno de ellos. *Boosting* se fundamenta en esta idea, realizando una combinación adaptativa de modelos simples (distintos entre sí) para dar lugar a un buen regresor. Utilizaremos concretamente el modelo **AdaBoost** visto en teoría.

La principal característica de *AdaBoost* es que ajusta una secuencia de modelos de aprendizaje débiles (es decir, modelos que mejoran ligeramente las suposiciones aleatorias, como los árboles de decisión) en versiones de los datos modificados repetidamente. Las predicciones de todos ellos se combinan mediante un voto mayoritario ponderado para producir la predicción final.

Los regresores simples utilizados van a ser considerados dentro de una clase simple, en nuestro caso una restricción de la clase de los  árboles de decisión $\mathcal{H}_{tree}$, donde limitamos de una forma muy acusada su crecimiento para se aprendan reglas muy generales y así forzar la simplicidad de los modelos. Por consiguiente, nuestra clase de funciones será:
\begin{equation}
\mathcal{H}_{ada\_boost} = \{h(x) = \sum_{t=1}^{T}α_th_t(x):h_t\in\mathcal{H}_{tree}\}. \end{equation}

Uno de los problemas que puede tener *AdaBoost* es que entrenar modelos muy simples de forma adaptativa pueden verse muy afectado por el ruido aleatorio relativo a los datos. Si empleásemos modelos más complejos, el sobreajuste podría aumentar muy rápidamente. Así, es prácticamente imposible desde nuestra posición medir estos errores de naturaleza aleatoria, con lo que surge el interés de probar otros *ensembles* que se sirvan de modelos más complejos, presentando por lo tanto una mayor robustez ante este tipo de ruido.

##### **Random Forest**

Los *Random Forests* también son ensembles basados en  ́arboles de decisión. Sin embargo, se sirven de tipos de técnicas distintas a las utilizadas en *AdaBoost*. La estructura de los  árboles no se limita tan acusadamente como ocurre con *AdaBoosting*, pero cada uno de los  árboles se entrena sólo con un subconjunto de datos muestreados con reemplazamiento sobre la muestra original (**bootstrap**). La predicción final, en nuestro caso de regresión, se realiza promediando los valores predichos por cada  árbol. Surgieron como forma de disminuir la varianza de la clase de funciones de los  árboles de decisión, aunque se pierda en cierta medida interpretabilidad del modelo obtenido.

La clase de funciones de los *Random Forests* es la siguiente:

\begin{equation} \mathcal{H}_{RF}=\{\frac{1}{N}\sum_{i=0}^{N}\sum_{j=0}^{M}α_{i,j}\mathbb{1}\chi_{i,j}(x):α_{i,j},\chi_{i,j} ~cumplen~ C^*\}, \end{equation}

donde las condiciones de $C^*$ son:

- $α_{i,j} \in \mathbb{R}$.
- $\chi_{i,j}=(a_{i,j},b_{i,j})×(c_{i,j},d_{i,j}):a_{i,j}<b_{i,j},c_{i,j}<d_{i,j} ∀i \in \{1,...,N\},j\in\{1,...,M\}.$
- Los conjuntos $\chi_{i,j}$ forman una partición de $X$. Es decir, $∀i\in\{1,...,N\}, \cup_{j=0}^{M}\chi_{i,j}=X$, y $\chi_{i,j}∩\chi_{i,l} = ∅ ∀j,l \in \{1,...,M\},i\in\{1,...N\}, j \neq l\}$.

Esta clase de funciones tiene en general una muy buena capacidad de adaptación y reporta muy buenos resultados. Con una cantidad suficiente de árboles se puede dar una predicción bastante precisa, y puesto que se devuelve un valor promediando los resultados devueltos por los  árboles entrenados con subconjuntos de características e instancias no es tan propenso a sobreajustar como otros modelos. Es por ello por lo que lo consideraremos y lo incluiremos entre nuestro conjunto de modelos empleados.

##### **Funciones de Base Radial**

Las Funciones de Base Radial (RBF) suponen otro paradigma de clases de funciones, basadas en el concepto del aprendizaje por semejanza, pero que de igual manera están muy estrechamente relacionadas con los modelos lineales. En nuestro caso, consideraremos RBF basadas en un núcleo gaussiano: $ϕ(z)=e^{\frac{-1}{2}z^2}$. La idea es fijar una serie de centros $\mu_j$ en los datos de entrada y considerarlos como referencia para aplicar el núcleo utilizando la distancia de cada punto a los centros $v_j$, reescalada por un parámetro $r_j$. La clase de estas funciones es la siguiente:

\begin{equation} \mathcal{H}_{RBF}=\{\sum_{j=1}^{k}w_jϕ(\frac{\lvert\lvert x-v_j\rvert\rvert}{r_j}):w_i,r_i\in \mathbb{R}, v_j \in \mathbb{R}^d, i\in \{0,...,k\}\}. \end{equation}

Como los centros son un parámetro del modelo y estos aparecen dentro de una función no lineal, este modelo es no lineal. Este modelo no está implementado en *Scikit-Learn* por lo que hemos tenido que crear la clase $\texttt{RBFNetworkRegressor}$, basada en el método dado por *V. Cherkassky and F. M. Mulier. Learning from Data: Concepts, Theory, and Methods. Wiley-IEEE Press, 2007* en la sección *Radial Basis Function Networks*. Si $X$ es el conjunto de datos, $y$ el conjunto de etiquetas, $k$ el número de centros y $γ\in[1,3]$ el radio de amplitud, los pasos del algoritmo son los siguientes:
1. $v=\{v_0,...,v_{k-1}\}=k\_medias(X,k)$.
2. $r_i = γmín_{j\neq i}\{\lvert\lvert v_i-v_j \rvert\rvert_2\}, i \in \{1,...,k-1\}$.
3. $r = \{r_0,...,r_{k-1}\}$.
4. $Z=KernelRBF(\frac{d(X,v)}{r})$.
5. $Ridge.fit(Z,y)$.

En la referencia anterior se escoge un radio $r_j$ para cada centro $v_j$, donde
$r_j = γmín_{k\neq j}\{\lvert\lvert v_k-v_j \rvert\rvert_2\}, γ∈[1,3]$.



```python
class RBFNetworkRegressor(BaseEstimator, RegressorMixin):
    """
       Regresor de red de funciones (gaussianas) de base radial.
       Uso de un modelo Ridge para ajustar los pesos del modelo final.
    """
    def __init__(self, k = 16, alpha = 1.0, batch_size = 128,
                 random_state = None, gamma=1):
        """
          Instancia un regresor con los hiperparámetros:
             - k:             número de centros a elegir.
             - alpha:         constante de regularización.
             - batch_size:    tamaño del batch para el clustering no supervisado.
             - random_state:  semilla aleatoria.
             - gamma:         Parámetro para controlar la escala de r
        """
        self.k = k
        self.alpha = alpha
        self.batch_size = batch_size
        self.random_state = random_state
        self.centers = None # Centroides
        self.r = None
        self.gamma = gamma
    def generar_centros_kmeans(self, X):
        """
          Elección de K centros a partir del algoritmo de K-Medias
        """
        init_size = 3 * self.k if 3 * self.batch_size <= self.k else None
        kmeans = MiniBatchKMeans(
            n_clusters = self.k,
            batch_size = self.batch_size,
            init_size = init_size,
            random_state = self.random_state)
        kmeans.fit(X)
        self.centers = kmeans.cluster_centers_

    def calcular_radio(self):
        """
          Cálculo del radio para aplicar en el kernel
        """
        # Cálculo la media de las distancias mínimas entre los centros
        mindists = np.zeros(self.centers.shape[1]) # Almacenamiento de las distancias entre cada par de CETOs
        for i in range(self.centers.shape[0]):
          min = float('inf')
          for j in range(self.centers.shape[0]):
            curdist = norm(self.centers[i]-self.centers[j])
            if curdist < min and not i == j:
              min = curdist
          mindists[i] = min
        self.r = self.gamma * mindists

    def _transform_rbf(self, X):
        """
          Aplicación del kernel RBF
        """
        ret = []
        for i in range(self.centers.shape[0]):
          ret.append(rbf_kernel(X, self.centers[i].reshape(-1,1).T , 1 / (2 * self.r[i] ** 2)))
        return np.array(ret).T.reshape(X.shape[0], self.centers.shape[0])

    def fit(self, X, y):
        """
          Entrena nuestro modelo
        """
        self.model = Ridge(alpha = self.alpha,
                           random_state = self.random_state)

        # Calculamos los k centros y el radio para el kernel
        self.generar_centros_kmeans(X)
        self.calcular_radio()
        # Transformamos los datos a partir de nuestro kernel y centros
        Z = self._transform_rbf(X)
        # Entrenamos...
        self.model.fit(Z, y)
        # Guardamos los coeficientes obtenidos
        self.intercept_ = self.model.intercept_
        self.coef_ = self.model.coef_
        return self

    def score(self, X, y = None):
        Z = self._transform_rbf(X)
        return self.model.score(Z, y)
    def predict(self, X):
        Z = self._transform_rbf(X)
        return self.model.predict(Z)
    def decision_function(self, X):
        Z = self._transform_rbf(X)
        return self.model.decision_function(Z)
```

##### **Support Vector Regressor**

Ahora cambiamos totalmente de paradigma. En los modelos SVM, y en particular en SVR, la regularización no va asociada a la norma de los pesos, sino a un término ligado con la anchura de la *tubería*. En particular, los parámetros libres que se emplean son ε y $C$. Se recuerda el problema primal que se resuelve en este problema ε-SVR: dados $x_i \in \mathbb{R}^d, i = 1,..., n$, siendo el kernel $\phi$ y las características $y \in \mathbb{R}^n$, el problema primal es:

\begin{equation}
\begin{split}
mín_{w,b,Ϛ,Ϛ^*}&\frac{1}{2}w^Tw+C\sum_{i=1}^{n}(Ϛ_i+Ϛ^*_i)\\
subject~to:&y_i-w^T\phi(x_i)-b \leq ε+Ϛ_i,\\
&w^T\phi(x_i)+b-y_i \leq ε+Ϛ^*_i,\\
&Ϛ_i,Ϛ^*_i \geq 0, i=1,...,n.
\end{split}
\end{equation}

Observamos que se penalizan las muestras tales que sus predicciones están al menos a distancia ε de su verdadera etiqueta. Estas muestras penalizan el objetivo a través de $Ϛ_i,Ϛ^*_i$, cuya suma está regulada por nuestro parámetro $\texttt{C}$. ε indica la anchura por defecto del *pasillo*. Estos han sido por tanto los dos parámetros empleados para llevar a cabo la regularización. Los espacios de valores considerados para ambos parámetros son los siguientes:
\begin{equation}
\begin{split}
\texttt{SVRCSpace} &= \{10^i : i \in \{−4, −3, −2\}\}\\
\texttt{SVRepsilonSpace} &= \{10^i : i \in \{−4, −3, −2\}\}.
\end{split}
\end{equation}

Hemos considerado que estos valores son los correctos para regularizar $\texttt{C}$ por tratarse de valores habituales para parámetros de regularización. Por otro lado, el espacio anterior para ε se ha visto adecuado debido a que regula la anchura de la *tubería* según unas unidades completamente coherentes respecto a las de las etiquetas, dejando un *pasillo* muy estrecho en el caso más extremo que permite que se lleguen a mejores soluciones. En *Scikit-Learn* el valor por defecto es 0.1, bastante elevado en nuestra opinión.

#### **Selección de la mejor hipótesis (Cross-Validation)**

Como ya comentamos en la sección correspondiente a la partición del conjunto de datos en datos de entrenamiento y de test, aplicaremos ahora **Cross-Validation 5-folds**. Lo haremos a través de la función $\texttt{GridSearchCV}$ de *Scikit-Learn*. Consideramos que Cross-Validation es el método de elección de un modelo más confiable, como se ha visto en teoría. Nótese que contamos con una gran variedad de estimadores posibles, cada uno de ellos presentando diversas opciones a nivel de ajuste de parámetros. El error de validación se torna representativo, pudiéndose tomar como un estimador de lo que sería el error fuera de la muestra.

La mejor hipótesis obtenida a través de este método ha sido $\texttt{RandomForestRegressor(max_depth=16, max_features=’sqrt’, n_estimators=200, n_jobs=-1)}$.

Con esta hipótesis se ha obtenido un error de validación cruzada $E_{CV} ≈ 3.250\cdot10^{−5}$ (a partir del MSE), lo que equivale a $0.0057 MW$ en términos del RMSE, un error bastante bajo que sostiene el buen ajuste realizado sobre los datos de entrenamiento.







```python
print(X_train.shape)
print(y_train.shape)
# Ejecutamos ahora Cross-Validation para hallar el mejor modelo
print("Selección del mejor modelo mediante Cross-Validation y el fit...")
start = default_timer()
pipe = Pipeline([('Model', AdaBoostRegressor())])

tiempo      = []
best_models = []
best_params = []
results     = []
for i in range(len(algs_and_params)):
  start = default_timer()
  best_model = GridSearchCV(pipe,
                algs_and_params[i],
                scoring=['neg_mean_squared_error',
                      'r2'],
                refit='neg_mean_squared_error',
                return_train_score=True,
                cv=5,
                n_jobs=-1,
                verbose=3)
  if i == 0:
    best_model.fit(X_train[:,:32], y_train)
  else:
    best_model.fit(X_train, y_train)
  tiempo.append(default_timer() - start)
  best_models.append(best_model)
  best_params.append(best_model.best_params_['Model'])
```

El *score* se ha visto transformado: en su momento comentamos que habíamos empleado el *score* $−ERM$, pero para representar los datos se ha decidido usar RMSE, debido a que permite una mejor interpretabilidad.

```python
def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
  Fuente:
    https://scikit-learn.org/stable/auto_examples/model_selection/plot_learning_curve.html#sphx-glr-auto-examples-model-selection-plot-learning-curve-py

    Generate 3 plots: the test and training learning curve, the training
    samples vs fit times curve, the fit times vs score curve.

    Parameters
    ----------
    estimator : estimator instance
        An estimator instance implementing fit and predict methods which
        will be cloned for each validation.

    title : str
        Title for the chart.

    X : array-like of shape (n_samples, n_features)
        Training vector, where ``n_samples`` is the number of samples and
        ``n_features`` is the number of features.

    y : array-like of shape (n_samples) or (n_samples, n_features)
        Target relative to ``X`` for classification or regression;
        None for unsupervised learning.

    axes : array-like of shape (3,), default=None
        Axes to use for plotting the curves.

    ylim : tuple of shape (2,), default=None
        Defines minimum and maximum y-values plotted, e.g. (ymin, ymax).

    cv : int, cross-validation generator or an iterable, default=None
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:

          - None, to use the default 5-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:CV splitter,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:StratifiedKFold used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:KFold is used.

        Refer :ref:User Guide <cross_validation> for the various
        cross-validators that can be used here.

    n_jobs : int or None, default=None
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:joblib.parallel_backend context.
        ``-1`` means using all processors. See :term:Glossary <n_jobs>
        for more details.

    train_sizes : array-like of shape (n_ticks,)
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the ``dtype`` is float, it is regarded
        as a fraction of the maximum size of the training set (that is
        determined by the selected validation method), i.e. it has to be within
        (0, 1]. Otherwise it is interpreted as absolute sizes of the training
        sets. Note that for classification the number of samples usually have
        to be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes,
                       return_times=True)

    # Convertimos los puntajes en relación al error cuadrático medio.
    train_scores = np.sqrt(-train_scores)
    test_scores = np.sqrt(-test_scores)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Graficamos las curvas de aprendizaje
    axes[0].grid()
    axes[0].fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes[0].fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes[0].plot(train_sizes, train_scores_mean, '*-', color="g",
                 label="Training score")
    axes[0].plot(train_sizes, test_scores_mean, '*-', color="b",
                 label="Cross-validation score")
    axes[0].legend(loc="best")

    # Graficamos los tiempos de ajuste en función del número de muestras
    axes[1].grid()
    axes[1].plot(train_sizes, fit_times_mean, '*-')
    axes[1].fill_between(train_sizes, fit_times_mean - fit_times_std,
                         fit_times_mean + fit_times_std, alpha=0.1)
    axes[1].set_xlabel("Training examples")
    axes[1].set_ylabel("fit_times")
    axes[1].set_title("Scalability of the model")

    # Graficamos la puntuación en función del tiempo de ajuste
    axes[2].grid()
    axes[2].plot(fit_times_mean, test_scores_mean, '*-')
    axes[2].fill_between(fit_times_mean, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1)
    axes[2].set_xlabel("fit_times")
    axes[2].set_ylabel("Score")
    axes[2].set_title("Performance of the model")

    return plt
```
